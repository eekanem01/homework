**Task:**

Write a function `total_length(words)` that returns the sum of the lengths of all words in a list using the accumulator pattern.

Write a second version using list comprehension.

**Sample Output:**

```
total_length(['data', 'science', 'rocks'])
```
should return `15`




`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def total_length(words):
  # initialize accumulator
  total = 0

  #for each word in the list
  for w in words:
    w_length = len(w) # get length
    total += w_length #

  return total
`````

`````python
# second version
def total_length(words):
  sum_word = [len(x) for x in words]
  total = sum(sum_word)
  return total
`````

**Task:**

Write a function `union(a, b)` that returns a new list containing all unique elements from both lists.

Rules:

* Cannot use a set

* Cannot modify input lists

* Must create a new list

* Order does not matter

**Sample Output:**

```

union of [4, 5] [1, 2, 3, 4] should return [4, 5, 1, 2, 3]
```




`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def union(a, b):
  # using info from stack overflow and list comprehension

  new_list = a + [x for x in b if x not in a]
  return new_list
`````

`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def union_2(a, b):
  # using info from stack overflow
  combined = a+b

  #converting to a set
  set_list = set(combined)

  return list(set_list) # return as a list
`````

**Problem 11**

**Concept:** *Using dictionary data type, loop, decision statements.*

**Task:**

Write a function `best_score(games, team_name)` that returns the game dictionary where the team scored the most points.

**Sample Output:**
```
games = [
    {'A': 8, 'B': 10},
    {'A': 15, 'C': 7},
    {'A': 12, 'D': 9}
]
print(best_score(games, 'A'))
```
should return
```
{'A': 15, 'C': 7}
```




`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def best_score(games, team_name):
  # get first game
  largest_game = games[0]

  #loop through to compare each score and save the game with the bigger score to the largest_game variable
  for game in games:
    if largest_game[team_name] < game[team_name]:
      largest_game = game
  return largest_game
`````

**Problem 12**

**Concept:** *Using list data type, dictionary data type, sorting.*

**Task:**

Write a function `best_word(words)` that returns the word with the highest Scrabble score.

**Sample Output:**
```
best_word(['dog', 'quiz', 'cat'])
```
should return `quiz`, because `quiz` has the highest scrabble score.




`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def best_word(words):
  #initialize an accumulator to compare each score to get the highest score
  high_score = 0

  #loop through, get each word score using the function word_score, compare and save highest score to high_score variable
  for word in words:
    score = word_score(word)
    if score > high_score:
      high_score = score
  return word # return word with highest score
`````
